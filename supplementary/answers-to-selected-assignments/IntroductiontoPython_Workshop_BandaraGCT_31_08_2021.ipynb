# **Matrix Operations in Python with Numpy**

### **Matrix Addition and Matrix Subtraction**

```python
import numpy as np

# Matrix addition
m1 = np.array([[1, 6, 5],
               [3, 4, 8],
               [1, 2, 3]]) 

m2 = np.array([[3, 4, 6],
               [5, 6, 7],
               [6, 8, 7]]) 
m3 = m1 + m2
print('m1 + m2 : \n', m3)
```

### **Matrix Multiplication**

```python
import numpy as np

# Matrix multiplication
m1 = np.array([[1,4,7],
                [2,5,8]])
m2 = np.array([[1,4],
                [2,5],
                [3,6]])
m3 = np.matmul(m1,m2) 
print('m1*m2: \n',m3) 


m4 = np.array([[1, 6, 5],
               [3, 4, 8],
               [1, 2, 3]]) 

m5 = np.array([[3, 4, 6],
               [5, 6, 7],
               [6, 8, 7]]) 
m6 = np.matmul(m4,m5) 
print('m4*m5: \n',m6) 
```

### **Transpose of a Matrix**

```python
import numpy as np

# Transpose of a Matrix
m1 = np.array([[1,4,7],
                [2,5,8]])
m2 = np.array([[1,4],
                [2,5],
                [3,6]])

m1_trs = m1.transpose()
print('Transpose of m1: \n', m1_trs)


m2_trs = m2.transpose()
print('Transpose of m2: \n', m2_trs)
```

## **Inverse of a matrix**

```python
import numpy as np

# Inverse of a matrix
m1 = np.array([[1, 0, 5],
               [2, 1, 6],
               [3, 4, 0]])


m1_inv = numpy.linalg.inv(m1)
print('Inverse of m1 : \n', m1_inv)

m2 = np.array([[2, -3, 4],
               [-3, -3, -2],
               [-2, 1, -1]])


m2_inv = numpy.linalg.inv(m2)
print('Inverse of m2 : \n', m2_inv)

```

## **Inner Product**

```python
import numpy as np

# Inner product
m1 = np.array([1,4,7])
m2 = np.array([1,0,2])

m_innerP = np.dot(m1, m2) 
print('Inner product of m1 and m2: \n', m_innerP)
```

## **Cross Product**

```python
import numpy as np

# Cross product
m1 = np.array([1,4,7])
m2 = np.array([1,0,2])

m_crossP = np.cross(m1, m2) 
print('Cross product of m1 and m2: \n', m_crossP)
```

## **Determinant of a Matrix**

```python
import numpy as np

# Determinant of a matrix
m1 = np.array([[1, 0, 5],
               [2, 1, 6],
               [3, 4, 0]])


m1_det =np.linalg.det(m1)
print('Det of m1 : \n', m1_det)

m2 = np.array([[1, 0, 0],
               [0, 2, 0],
               [0, 0, 4]])


m2_det = np.linalg.det(m2)
print('Det of m2 : \n', m2_det)
```

## Eigenvalues and Eigen Vectors of a Matrix

```python
import numpy as np

# Eigenvalues and eigenvectors of a matrix
m1 = np.array([[2, 1, -2],
               [1, 0, 0],
               [0, 1, 0]])


m1_eValues, m1_eVectors =np.linalg.eig(m1)
print('Eigenvalues of m1 : \n', m1_eValues)
print('Eigenvectors of m1 : \n', m1_eVectors)

m2 = np.array([[1, 0, 0],
               [0, 2, 0],
               [0, 0, 4]])

m2_eValues, m2_eVectors =np.linalg.eig(m2)
print('Eigenvalues of m2 : \n', m2_eValues)
print('Eigenvectors of m2 : \n', m2_eVectors)
```
